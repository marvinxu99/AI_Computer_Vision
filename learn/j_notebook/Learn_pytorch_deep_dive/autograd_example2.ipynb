```python
import torch
import matplotlib.pyplot as plt

# Define input tensor
a = torch.linspace(-3, 3, 100, requires_grad=True)

# Define function
out = a**3 + 2*a**2 - 5*a + 1  # f(a)

# Compute gradient
out.backward(torch.ones_like(a))

# Print gradient
print(a.grad)

# Plot gradient vs. input
plt.plot(a.detach(), a.grad.detach(), label="f'(a)")
plt.title("Gradient of f(a)")
plt.xlabel("a")
plt.ylabel("df/da")
plt.legend()
plt.grid(True)
plt.show()
```
