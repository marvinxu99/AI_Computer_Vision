# Full implementation of training a 2-layer Neural Network

```python
import numpy as np
from numpy.random import randn

import matplotlib.pyplot as plt
```

```python
# Weight initialization case: values are too small

dims = [4096] * 7
hs = [ ]
x = np.random.randn(16, dims[0])

plt.figure(figsize=(22, 4))

# Forward pass with ReLU activation
for i, (Din, Dout) in enumerate(zip(dims[:-1], dims[1:])):
    W = 0.01 * np.random.randn(Din, Dout)                    # 0.01
    x = np.maximum(0, x.dot(W))
    hs.append(x)

    vals = x.flatten()
    mean = vals.mean()
    std  = vals.std()

    plt.subplot(1, 7, i+1)
    plt.hist(vals, bins=150)
    plt.title(f"Layer {i+1}\nmean={mean:.02f}, std={std:.02f}")
    plt.xlim([-1, 1])  # same x-axis
    plt.xticks([-1, 0, 1])

plt.tight_layout()
plt.show()
```

```python
# Weight initialization case: values are too large

dims = [4096] * 7
hs = [ ]
x = np.random.randn(16, dims[0])

plt.figure(figsize=(22, 4))

# Forward pass with ReLU activation
for i, (Din, Dout) in enumerate(zip(dims[:-1], dims[1:])):
    W = 0.05 * np.random.randn(Din, Dout)                       # 0.05
    x = np.maximum(0, x.dot(W))
    hs.append(x)

    vals = x.flatten()
    mean = vals.mean()
    std  = vals.std()

    plt.subplot(1, 7, i+1)
    plt.hist(vals, bins=150)
    plt.title(f"Layer {i+1}\nmean={mean:.02f}, std={std:.02f}")
    plt.xlim([-1, 1])  # same x-axis
    plt.xticks([-1, 0, 1])

plt.tight_layout()
plt.show()
```

```python
# Weight initialization case: values are right 

dims = [4096] * 7
hs = [ ]
x = np.random.randn(16, dims[0])

plt.figure(figsize=(22, 4))

# Forward pass with ReLU activation
for i, (Din, Dout) in enumerate(zip(dims[:-1], dims[1:])):
    W = np.random.randn(Din, Dout) * np.sqrt(2.0 / Din)        # He init    
    x = np.maximum(0, x.dot(W))
    hs.append(x)

    vals = x.flatten()
    mean = vals.mean()
    std  = vals.std()

    plt.subplot(1, 7, i+1)
    plt.hist(vals, bins=150)
    plt.title(f"Layer {i+1}\nmean={mean:.02f}, std={std:.02f}")
    plt.xlim([-1, 1])  # same x-axis
    plt.xticks([-1, 0, 1])

plt.tight_layout()
plt.show()
```
